# Data Cleaning and Wrangling Phase

```python
hrDummies = pd.get_dummies(hrData,columns=categorical)
hrDummies.describe()
```

```python
hrDummies
```

```python
hrDummies.drop(['EmployeeCount','StandardHours','Over18_Y','Attrition_No','EmployeeID'],axis=1, inplace=True)
```

```python
hrDummies.info()
```

```python
hrDummies['Attrition_Yes'].value_counts()
```

```python
hrDummies.isnull().sum()
```

```python
hrDummies['NumCompaniesWorked'].fillna(0.0,inplace=True)
hrDummies['TotalWorkingYears'].fillna(0.0,inplace=True)
```

# Scaling the data(Standarize)

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
attrition_data = hrDummies.drop(['Attrition_Yes'],axis = 1)
attrition_label = hrDummies['Attrition_Yes']
scaler = preprocessing.StandardScaler()
scaled_att_data = scaler.fit_transform(attrition_data)
```

```python
names = attrition_data.columns
scaled_attrition_data = pd.DataFrame(scaled_att_data, columns=names)
```

```python
data_train_scale, data_test_scale, label_train_scale, label_test_scale = train_test_split(scaled_attrition_data, attrition_label, test_size=0.20)
```

# Model Building-Evaluating Phase

```python
import tensorflow as tf
import keras  
from keras import layers
from keras import losses
```

```python
model = keras.Sequential([
    keras.layers.Dense(45, activation='relu',kernel_initializer=keras.initializers.he_uniform(),bias_initializer='zeros',input_shape=(39,)),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(12, activation='relu',kernel_initializer=keras.initializers.he_uniform(),bias_initializer='zeros'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(6, activation='relu',kernel_initializer=keras.initializers.he_uniform(),bias_initializer = "zeros"),
    keras.layers.Dense(1, activation='sigmoid') ])
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(0.001),
              loss='binary_crossentropy',
              metrics=['accuracy']) 
```

```python
 model.fit(data_train_scale,label_train_scale,batch_size = 100,epochs =200,verbose = 2)
```

```python
test_loss ,test_acc = model.evaluate(data_test_scale,label_test_scale)
print('Test accuracy:', test_acc, test_loss)
```
